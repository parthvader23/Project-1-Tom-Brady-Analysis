```python
import pandas as pd
```

```python
championship_csv = "Championship_team_stats.csv"
qb_csv = "Championship_QB_data.csv"
qb_csv
```

```python
champ_df=pd.read_csv(championship_csv)
qb_df=pd.read_csv(qb_csv)
# champ_df
# qb_df
```

```python
championship_stats_df=pd.merge(champ_df,qb_df, on="Result",how="outer")
```

```python
championship_stats_df.columns
```

```python
champ_df=championship_stats_df.drop(['Tm_y', 'Year?', 'Unnamed: 2_y', 'Opp_y','OT_y'],axis=1)
```

```python
champ_stats=championship_stats_df.rename(columns=
                                        {'Tm_x':'Teamn',
                                        'Unnamed: 2_x':'@',
                                        'Opp_x':'Opponent',
                                        'OT_x':'Overtime'
                                        })
```

```python
champ_df=champ_stats.drop(['Overtime','Tm_y', 'Year?', 'Unnamed: 2_y', 'Opp_y','OT_y'],axis=1)
```

```python
champ_df=champ_df.fillna(0)
```
